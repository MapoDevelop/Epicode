**Quandrato di un numero**

Creo un programma che dato un nuomero dall'utente, stampi il 
*risultato* del quadrato di quel numero

```python
numero = int(input("Di che numero vuoi che ti faccia il quadrato?"))
quadrato = int(pow(numero, 2)) 
print(f"Il quadrato di {numero} è {quadrato}")
```
